```python
if __name__ == "__main__":
    # Example usage (typically this would be in a separate script or notebook)
    
    # 1. Initialize Database (interactive)
    db_path = search_and_initialise_db() # Searches parent directory of script
    if db_path is None:
        print("Database not initialized. Exiting.")
        sys.exit()
    print(f"Using database: {qc.config.core.db_location}")

    # 2. Example: Load a dataset and get info (replace with an actual run_id)
    example_run_id = 1 # Replace with a valid run_id from your database
    try:
        print(f"\n--- Info for run_id {example_run_id} ---")
        # Unpack start_time from get_dataset_info as it's used in plot_heatmaps
        exp_name_ex, sample_name_ex, param_units_ex, start_time_ex, completed_time_ex, duration_ex, display_time_ex = get_dataset_info(example_run_id)
        setpoints, dep, p_info, data_df = info_df(example_run_id) # info_df also calls get_dataset_info internally via cached_load_dataset
        
        # 3. Example: Plotting (if data is suitable for heatmaps)
        if len(setpoints) >= 2:
            print(f"\n--- Plotting heatmaps for run_id {example_run_id} ---")
            fig_v, fig_dvdi, fig_ic, df_ic = plot_heatmaps(example_run_id) # plot_heatmaps calls get_dataset_info again
            if fig_v:
                fig_v.show()
                # plot_save(fig_v, f"run{example_run_id}_voltage_{sample_name_ex}", save_dir="plots", save_png=True, save_pdf=False)
            if fig_dvdi:
                fig_dvdi.show()
                # plot_save(fig_dvdi, f"run{example_run_id}_dvdi_{sample_name_ex}", save_dir="plots", save_png=True, save_pdf=False)
            if fig_ic:
                fig_ic.show()
                # plot_save(fig_ic, f"run{example_run_id}_Ic_{sample_name_ex}", save_dir="plots", save_png=True, save_pdf=False)
            if df_ic is not None:
                print("\n--- Estimated Ic DataFrame ---")
                print(df_ic.head())
        elif len(setpoints) == 1 :
             print(f"\n--- Plotting 1D data for run_id {example_run_id} ---")
             fig_1d, _, _, _ = plot_heatmaps(example_run_id) # plot_heatmaps handles 1D case
             if fig_1d:
                 fig_1d.show()
                 # plot_save(fig_1d, f"run{example_run_id}_1D_{sample_name_ex}", save_dir="plots", save_png=True, save_pdf=False)
        else:
            print(f"Dataset {example_run_id} has no setpoints for plotting.")


        # 4. Example: Fitting (using some dummy data or data from df)
        if not data_df.empty and len(data_df) > 2 and setpoints: # Ensure setpoints list is not empty
            print("\n--- Example Fitting ---")
            # Assuming first setpoint and dependent parameter are suitable for a linear fit
            x_param_name_fit = setpoints[0]
            y_param_name_fit = dep
            
            if x_param_name_fit in data_df.columns and y_param_name_fit in data_df.columns:
                x_data_fit = data_df[x_param_name_fit].values
                y_data_fit = data_df[y_param_name_fit].values
                
                # Ensure no NaNs for fitting
                valid_indices = ~np.isnan(x_data_fit) & ~np.isnan(y_data_fit)
                x_data_fit = x_data_fit[valid_indices]
                y_data_fit = y_data_fit[valid_indices]

                if len(x_data_fit) > 2:
                    # Get units for fitting printout
                    x_unit_fit = ureg(param_units_ex.get(x_param_name_fit, ""))
                    y_unit_fit = ureg(param_units_ex.get(y_param_name_fit, ""))

                    lin_fit_res = polyfit("Linear Fit Example", x_data_fit, y_data_fit)
                    print_fit_result(lin_fit_res, x_unit=x_unit_fit, y_unit=y_unit_fit)

                    const_fit_res = constfit("Constant Fit Example", x_data_fit, y_data_fit)
                    print_fit_result(const_fit_res, x_unit=x_unit_fit, y_unit=y_unit_fit)
                else:
                    print("Not enough valid data points for example fitting after NaN removal.")
            else:
                print(f"Columns for fitting ('{x_param_name_fit}', '{y_param_name_fit}') not found in DataFrame.")
        else:
            print(f"DataFrame for run_id {example_run_id} is empty, too small, or has no setpoints for fitting example.")

    except FileNotFoundError:
        print(f"Error: Database file not found for run_id {example_run_id}. Please check DB path and `search_and_initialise_db`.")
    except ValueError as ve:
        print(f"ValueError during example for run_id {example_run_id}: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred during example usage for run_id {example_run_id}: {e}")
        print("Please ensure the run_id is valid and the database is accessible.")

    # Example for generate_points_for_parameter
    print("\n--- Example: generate_points_for_parameter ---")
    points = generate_points_for_parameter(
        MagCenter=0.0,      # Tesla
        RangeOuter=0.1,     # Tesla
        RangeInner=0.05,    # Tesla
        RangeCenter=0.01,   # Tesla
        step_outer=0.01,    # Tesla
        step_inner=0.002,   # Tesla
        step_center=0.0005  # Tesla
    )
    print(f"Generated {len(points)} points for parameter sweep:")
    # print(points) # Can be very long
    if len(points) > 0:
        print(f"Min: {points.min():.4f} T, Max: {points.max():.4f} T")
    else:
        print("No points generated.")

```
